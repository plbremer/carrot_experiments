```python

#we COULD get this list from everything that we have, but for now we choose a hardcoded list seen at the top of the notebook
#unique_species_organ_pairs=list(set(temp['species_and_organ'].to_list()))
unique_species_organ_pairs=species_organ_pairs_with_count_20
```

```python
sub_panda_list=[temp.loc[temp['species_and_organ']==unique_species_organ_pairs[i]].copy() for i in range(len(unique_species_organ_pairs)) ]
```

```python
#[sub_panda_list_element for sub_panda_list_element in sub_panda_list]
```

```python
len(unique_species_organ_pairs)
```

```python
for i in range(len(sub_panda_list)):
    sub_panda_list[i].drop(['species','organ','species_and_organ'],inplace=True,axis='columns')
    sub_panda_list[i]=sub_panda_list[i].pivot(columns='name',index='bin',values='count')    
```

```python
indices_that_each_panda_needs=[float(i) for i in range(0,my_intensity_max,25000)]
```

```python
import numpy as np
```

```python
temp_to_append=pandas.DataFrame(index=indices_that_each_panda_needs,columns=sub_panda_list[0].columns.to_list(),data=np.nan)
temp_to_append
for i in range(len(sub_panda_list)):
    #print(i)
    sub_panda_list[i]=sub_panda_list[i].append(temp_to_append,ignore_index=False,verify_integrity=False)
    #sub_panda_list[i].drop_duplicates(keep='first',inplace=True)
    sub_panda_list[i]=sub_panda_list[i][~sub_panda_list[i].index.duplicated(keep='first')]
    #print(sub_panda_list[i])
```
